```python
import pandas as pd
from datetime import datetime
import yfinance as yf
import time
```

#### S&P500 Company list

```python
wikipedia_data = pd.read_html('https://en.wikipedia.org/wiki/List_of_S%26P_500_companies')
```

```python
snp_company_list = wikipedia_data[0] # the first table
snp_changes = wikipedia_data[1] # the second table
```

```python
snp_company_list.head()
```

```python
ticker_list = snp_company_list['Symbol'].unique().tolist()
len(ticker_list)
```

```python
snp_changes.head()
```

#### interval of 1 day

```python
# Define a start date and End Date
start = '2013-01-01'
#setting today date as End Date
end = datetime.today().strftime('%Y-%m-%d')
```

```python
data_list1=[]

t_begin = time.time()

for ticker in ticker_list:
    
    df = yf.download(tickers=ticker, start= start , 
                  end = end, interval = '1d').reset_index()
    
    df['Symbol'] = ticker
    data_list1.append(df)

t_end = time.time()
print(f"time elapsed in mins: {(t_end-t_begin)/60}")
```

```python
stock_1d = pd.concat(data_list1,axis=0)
stock_1d = stock_1d.merge(snp_company_list,on='Symbol',how='left')
print(stock_1d.shape)
stock_1d.head()
```

```python
stock_1d.info()
```

#### interval of 1 minute

```python
# Define a start date and End Date
start3 = '2023-10-10'
end3 = '2023-10-16'""

# Define a start date and End Date
start4 = '2023-10-16'
end4 = '2023-10-23'

# Define a start date and End Date
start5 = '2023-10-23'
end5 = '2023-10-30'


# Define a start date and End Date
start6 = '2023-10-30'
end6 = '2023-11-05'
```
